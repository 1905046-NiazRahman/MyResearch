```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import csv
```

```python
output_files = sorted(glob.glob('./output/*.twc'))
print(len(output_files))
```

```python
# Separate lists to store extracted data as integers
n_values = []
m_values = []
width_values = []
time_values = []
nbags_values = []

# Specify the directory containing the .twc files
output_directory = 'output'

# Iterate through each file in the output directory
for filename in os.listdir(output_directory):
    if filename.endswith('.twc'):  # Process only .twc files
        file_path = os.path.join(output_directory, filename)  # Create full path to the file
        with open(file_path, 'r') as file:
            lines = file.readlines()
            if len(lines) >= 4:  # Ensure there are at least 4 lines
                # Extracting values using string manipulation
                first_line = lines[0]
                width_time_line = lines[2]
                nbags_line = lines[3]

                # Extracting n and m from the first line
                n_value = int(first_line.split('n ')[1].split(' ')[0])  # Convert to integer
                m_value = int(first_line.split('m ')[1].split(' ')[0])  # Convert to integer

                # Extracting width and time from the third line
                width_value = int(width_time_line.split('width ')[1].split(' ')[0])  # Convert to integer
                time_value = int(width_time_line.split('time(ms) ')[1].strip())  # Convert to integer

                # Extracting nbags from the fourth line
                nbags_value = int(nbags_line.split('nbags ')[1].strip())  # Convert to integer

                # Append extracted values to their respective lists
                n_values.append(n_value)
                m_values.append(m_value)
                width_values.append(width_value)
                time_values.append(time_value)  # Cleaned time values
                nbags_values.append(nbags_value)

# Print the extracted data lists
print("N values:", n_values)
print("M values:", m_values)
print("Width values:", width_values)
print("Time values:", time_values)  # Cleaned time values
print("Nbags values:", nbags_values)
```

```python
# Combine n_values and width_values into a single array for sorting
combined = list(zip(n_values, width_values))

# Sort the combined list based on n_values
sorted_combined = sorted(combined)

# Unzip the sorted list back into two separate lists
sorted_n_values, sorted_width_values = zip(*sorted_combined)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(sorted_n_values, sorted_width_values, marker='o', linestyle='-')
plt.xlabel('No. of Nodes')
plt.ylabel('Treewidth')
plt.title('RTW Algorithm: Treewidth vs. Number of Nodes')
plt.grid(True)
plt.show()
```

```python
# Calculate graph density
density_values = [(2 * m) / (n * (n - 1)) if n > 1 else 0 for n, m in zip(n_values, m_values)]
```

```python
# Combine density and width_values into a single array for sorting
combined_density = list(zip(density_values, width_values))

# Sort the combined list based on density
sorted_combined_density = sorted(combined_density)

# Unzip the sorted list back into two separate lists
sorted_density_values, sorted_width_values = zip(*sorted_combined_density)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(sorted_density_values, sorted_width_values, marker='o', linestyle='-')
plt.xlabel('Graph Density')
plt.ylabel('Treewidth')
plt.title('RTW Algorithm: Treewidth vs. Graph Density')
plt.grid(True)
plt.show()
```

```python
# Combine density with time and bags for sorting
combined_density_time = list(zip(density_values, time_values))
combined_density_bags = list(zip(density_values, nbags_values))

# Sort the combined lists based on density
sorted_combined_time = sorted(combined_density_time)
sorted_combined_bags = sorted(combined_density_bags)

# Unzip the sorted lists back into separate lists
sorted_density_values_time, sorted_time_values = zip(*sorted_combined_time)
sorted_density_values_bags, sorted_nbags_values = zip(*sorted_combined_bags)

```
